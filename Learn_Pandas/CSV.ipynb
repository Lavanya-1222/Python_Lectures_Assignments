```python
pd.read_csv('file.tsv',sep='\t',names=['id','votes','rating','gross'])
```

```python
pd.read_csv('file.tsv',sep='\t',names=['id','votes','rating','gross'],index_col=0)
```

```python
# header if you want to mak
pd.read_csv('file.tsv',sep='\t',header=1)
```

```python
#usecols:used to only fetch needed columns
pd.read_csv('file.tsv',sep='\t',usecols=['0','50'])
```

```python
# skiprows -used to skip specified rows
pd.read_csv('file.tsv',sep='\t',skiprows=[4,2])
```

```python
# nrows only retriving specified rows
pd.read_csv('file.tsv',sep='\t',nrows=100)
```

```python
# dtype=if you want to change data type of any columns as time of runing
pd.read_csv('file.tsv',sep='\t',dtype={'5':float})
```

```python
# parse_date=[columns name]=is used to convert data type to data rather than object
```

```python
# convertors are used to apply function on specified columns
def g(x):
    if int(x)>4:
        return 'greater'
pd.read_csv('file.tsv',sep='\t',dtype={'5':str},converters={'5':g})
```

```python
# na_vales if you want to make any specified values as na values 
pd.read_csv('file.tsv',sep='\t',na_values=[0])
```

```python
# chunksize used to load data in chunks 
dfs=pd.read_csv('file.tsv',sep='\t',chunksize=5000)
for i in dfs:
    print(i.shape)
```
