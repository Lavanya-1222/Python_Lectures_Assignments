# Data Cleaning

```python
sp.info()
```

```python
# drop last 3 column
sp.drop(columns=['Unnamed: 3','Unnamed: 4','Unnamed: 2'],inplace=True)
```

```python
sp.sample()
```

```python
# Renameing Columns
sp.rename(columns={'v1':'target','v2':'test'},inplace=True)
```

```python
sp.sample()
```

```python
sp['target']=sp['target'].map({'ham':0,'spam':1})
```

```python
sp.head()
```

```python
# Check Missing Values
sp.isnull().sum()
```

```python
# check for duplicate values
sp.duplicated().sum()
```

```python
sp=sp.drop_duplicates(keep='first')
sp.duplicated().sum()
```

```python
sp.shape
```

```python
# we are going to find number of words ,charecters ans sentances in text
sp['Num_Charecters']=sp['test'].apply(lambda x : len(x))
```

```python
sp.head()
```

```python
sp['num_words']=sp['test'].apply(lambda x: len(x.split(" ")))
```

```python
sp.head()
```

```python
sp['num_sent']=sp['test'].apply(lambda x:len(x.split(',')))
```

```python
sp.head()
```

```python
sp[['Num_Charecters','num_words','num_sent']].describe()
```

```python
#ham
sp[sp['target']==0][['Num_Charecters','num_words','num_sent']].describe()
```

```python
#spam
sp[sp['target']==1][['Num_Charecters','num_words','num_sent']].describe()
```

```python
# num of charectors in ham and spam
sns.histplot(sp[sp['target']==0]['Num_Charecters'],color='Green')
sns.histplot(sp[sp['target']==1]['Num_Charecters'],color='red')
```

```python
# num of words in ham and spam
sns.histplot(sp[sp['target']==0]['num_words'],color='Green')
sns.histplot(sp[sp['target']==1]['num_words'],color='red')
```

```python
# num of sentances in ham and spam
sns.histplot(sp[sp['target']==0]['num_sent'],color='Green')
sns.histplot(sp[sp['target']==1]['num_sent'],color='red')
```

```python
# check relation ship of each graph with each other
sns.pairplot(sp,hue='target')
```

```python
# !pip install nltk
# nltk.download('punkt')
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
def transform(text):
    text=text.lower()
    text=word_tokenize(text)
    y=[]
    for i in text:
        if i.isalnum():
            y.append(i)
    return y
    
```
