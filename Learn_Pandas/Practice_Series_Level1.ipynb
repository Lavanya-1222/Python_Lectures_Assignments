```python
# 3. Create a Pandas Series with values [5, 10, 15, 20, 25] and set the index labels as ['A', 'B', 'C', 'D', 'E'].  
values=[5, 10, 15, 20, 25]
s=pd.Series(values,index=['A', 'B', 'C', 'D', 'E'])
s
```

```python
# 6. Find the sum, mean, and standard deviation of a Pandas Series containing [2, 4, 6, 8, 10].
s=pd.Series([2, 4, 6, 8, 10])
print(s.sum())
print(s.mean())
print(s.std())
```

# Merging

```python
#merging two dataframe
x=pd.DataFrame({'A':[1,2,3,4,5],'B':[6,7,8,9,10]})
y=pd.DataFrame({'A':[1,2,3,4,5],'C':[16,17,18,19,20]})
pd.merge(x,y,on='A')
```

```python
#merging two dataframe right
x=pd.DataFrame({'A':[1,2,3,4,5],'B':[6,7,8,9,10]})
y=pd.DataFrame({'A':[1,2,3,6,8],'C':[16,17,18,19,20]})
pd.merge(x,y,how='right')
```

```python
#merging two dataframe left
x=pd.DataFrame({'A':[1,2,3,4,5],'B':[6,7,8,9,10]})
y=pd.DataFrame({'A':[1,2,3,6,8],'C':[16,17,18,19,20]})
pd.merge(x,y,how='left')
```
