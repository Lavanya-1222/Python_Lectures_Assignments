```python
import matplotlib.pyplot as plt
import pandas as pd
```

 # 2D Line plot
 - Bivariate Analysis
 - categorical -> numerical and numerical -> numerical
 - Use case - Time series data (means increaseing or decrising data by time)

```python
price = [48000,54000,57000,49000,47000,45000]
year = [2015,2016,2017,2018,2019,2020]
plt.plot(year,price)
```

```python
df=pd.read_csv('sharma-kohli.csv')
df
plt.plot(df['index'],df['V Kohli'])
```

```python
# Ploting Multiple Plots
plt.plot(df['index'],df['V Kohli'])
plt.plot(df['index'],df['RG Sharma'])
```

```python
# labels titles
plt.plot(df['index'],df['V Kohli'])
plt.plot(df['index'],df['RG Sharma'])
plt.title('Rohith Sharma vs Virat Kohali')
plt.ylabel('Runs_Scored')
plt.xlabel('Season')
```

```python
# colors(hex)
plt.plot(df['index'],df['V Kohli'],color='yellow')
plt.plot(df['index'],df['RG Sharma'],color='green')
```

```python
# line(width and style)
plt.plot(df['index'],df['V Kohli'],linestyle='dashed')
plt.plot(df['index'],df['RG Sharma'],linestyle='dashdot',linewidth=3)
# Using linestyle Argument:
# Solid
# Dashed
# Dotted
# Dashdot
# None
```

```python
# marker & marker size
# o*.,X+PsDd
plt.plot(df['index'],df['V Kohli'],marker='o',markersize=8)
plt.plot(df['index'],df['RG Sharma'],marker='D')
```

```python
# legend->location
# 'upper left'
# 'upper right'
# 'lower left'
# 'lower right'
# 'upper center'
# 'lower center'
# 'center left'
# 'center right'
plt.plot(df['index'],df['V Kohli'],label='Virat')
plt.plot(df['index'],df['RG Sharma'],label='Sharma')
plt.legend()
plt.legend(loc='upper right')
```

```python
# limiting axes
price = [48000,54000,57000,49000,47000,45000,4500000]
year = [2015,2016,2017,2018,2019,2020,2021]

plt.plot(year,price)
plt.ylim(0,75000)
plt.xlim(2017,2019)
```

```python
# grid()
price = [48000,54000,57000,49000,47000,45000,4500000]
year = [2015,2016,2017,2018,2019,2020,2021]

plt.plot(year,price)
plt.grid()
```

```python
# show()- if your excuting this code in another place then the graph is not visible until u call plt.show() function

price = [48000,54000,57000,49000,47000,45000,4500000]
year = [2015,2016,2017,2018,2019,2020,2021]
plt.plot(year,price)
plt.show()
```

# Scatter Plot
 - Bivarinet Analysis
 - Numerical vs Numerical
 - Use Case: used for finding corelation

```python
df1=pd.read_csv('batter.csv')
df2=df1.head(50)
df2
```

```python
# plt.scatter # you can apply all function seen in previus learning like marker lable title
plt.scatter(df2['avg'],df2['strike_rate'])
```

```python
import seaborn as se
d=se.load_dataset('tips')
d
```

```python
# size
plt.scatter(d['total_bill'],d['tip'],s=d['size']*15)
```

```python
# scatterplot using plt.plot
plt.plot(d['total_bill'],d['tip'],'o')
```

```python
# plt.plot vs plt.scatter
# plot is faster that scatter
```

# Bar Chart
  - Bivarient Analysis
  - numerical vs chategorical
  - Use Case: aggregation Analysis of Groups

```python
# simple plot
childern=[10,20,40,10,30]
color=['red','blue','green','yellow','pink']
plt.bar(color,childern)
```

```python
# horizontal bar chart
childern=[10,20,40,10,30]
color=['red','blue','green','yellow','pink']
plt.barh(color,childern)
# if number of chtegories are increased then you can use horizontal barchart
```
