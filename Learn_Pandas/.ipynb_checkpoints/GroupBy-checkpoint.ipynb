```python
import pandas as pd
```

```python
df=pd.read_csv('imdb-top-1000.csv')
df
```

```python
df.groupby('Genre')
```

```python
# Find top 3 generas by total Earning
df.groupby('Genre')['Gross'].sum().sort_values(ascending=False).head(3)
```

```python
# find the genre with highest IMDB Rating
df.groupby('Genre')['IMDB_Rating'].mean().sort_values(ascending=False).head(1)
```

```python
# find no of movies done by each actor
df['Star1'].value_counts()
df.groupby('Star1')['Series_Title'].count()
```

```python
df.head(2)
directers=df.groupby('Director')
```

```python
# find director with most popularity
directers['No_of_Votes'].sum().sort_values(ascending=False).head(1)
```

```python
# find the highest rated movie of each genre
geners=df.groupby('Genre')
geners['IMDB_Rating'].max()
```

```python
# find no.of movies done by each charecter
df.groupby('Star1')['IMDB_Rating'].count().sort_values(ascending=False)
```

```python
# GroupBy Attributes and Methods
# find total number of groups -> len
# find items in each group -> size
# first()->first item/last()->last item ->/ nth(index)->item
# get_group -> vs filtering
# groups
# describe
# sample
# nunique
```

```python
# len-return number of categories in group
len(df.groupby('Genre'))
```

```python
# size- return items in each category
df.groupby('Genre').size()
```

```python
df.groupby('Genre').first()
df.groupby('Genre').last()
df.groupby('Genre').nth(6)
```

```python
# get_group-> return the table of specified category
df.groupby('Genre').get_group('Action')
```

```python
# describe ->return numerical column description in each category
df.groupby('Genre').describe()
```

```python
# sample-> return sample records from each category
df.groupby('Genre').sample()
```

```python
# nunique()->returns no of unique values in each category in each column
df.groupby('Genre').nunique()
```

```python
# agg method->  different aggregate function on diffrenent columns
# passing dict
df.groupby('Genre').agg(
    {
        'IMDB_Rating':'mean',
        'No_of_Votes':'sum',
        'Gross':'sum',
       'Metascore':'min'}
)
        
```

```python
# passing list
df.groupby('Genre').agg(
    ['min','max','sum'])
```

```python
# Adding Both Syntaxs
df.groupby('Genre').agg(
    {
        'Runtime':['min','mean'],
        'IMDB_Rating':'mean',
        'No_of_Votes':['sum','max'],
        'Gross':'sum',
        'Metascore':'min'
    }
)
```

```python
genrs=df.groupby('Genre')
```

```python
# looping on group
for group,data in genrs:
    print(group)
    print(data)
    
```

```python
# apply on group
genrs.apply(min)
```

```python
# find number of movies starting with A for each group
def starts(group):
    return group['Series_Title'].str.startswith('A').sum()
        
genrs.apply(starts)
```

```python
# find ranking of each movie in the group according to IMDB score
def rank_movie(group):
    group['Rank_By_IMDM']=group['IMDB_Rating'].rank(ascending=False)
    return group
genrs.apply(rank_movie)
```

```python
def normal(group):
    group['normal']=((group['IMDB_Rating']-group['IMDB_Rating'].min())/(group['IMDB_Rating'].max()-group['IMDB_Rating'].min()))
    return group
genrs.apply(normal)
```

```python
# find the most earning actor->director combo
df.groupby(['Director','Star1'])['Gross'].sum().sort_values(ascending=False).head(1)
```

```python
# find the best(in-terms of metascore(avg)) actor->genre combo
df.groupby(['Genre','Star1'])['Metascore'].mean().reset_index().sort_values('Metascore',ascending=False).head(1)
```

```python
# agg on multiple column
duo = df.groupby(['Director','Star1'])
duo.agg(['min','max','mean'])
```
