```python
import numpy as np
import pandas as pd
```

# Series from lists

```python
# string
country = ['India','Pakistan','USA','Nepal','Srilanka']

pd.Series(country)
```

```python
# integers
runs = [13,24,56,78,100]
pd.Series(runs)
```

```python
# custom index
marks = [67,57,89,100]
subjects = ['maths','english','science','hindi']
pd.Series(marks,index=subjects)
```

```python
# setting a name
marks = pd.Series(marks,index=subjects,name="lavanya marks")
marks
```

# Series from Dictionary

```python
marks = {
    'maths':67,
    'english':57,
    'science':89,
    'hindi':100
}
marks_series=pd.Series(marks,name="lavanya marks")
marks_series
```
