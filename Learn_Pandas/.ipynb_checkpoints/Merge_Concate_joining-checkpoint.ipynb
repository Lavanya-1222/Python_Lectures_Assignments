# Concat()
#### - Horizontally
#### - verticaly
#### - ignore_index
#### - Multiindexing
->

```python
# concating Horizontally
pd.concat([df1,stud2])
```

```python
# ignore_index
pd.concat([df1,stud2],ignore_index=True)
```

```python
# Multiindexing
t=pd.concat([df1,stud2],keys=['Table1','Table2'])
t
```

```python
t.loc[('Table2',5)]
```

```python
#concating verticle
pd.concat([df1,stud2],axis=1)
```

# Joining/Mergeing
- Left
- inner
- right
- outer

```python
df2=pd.read_csv('reg-month1.csv')
df4=pd.read_csv('reg-month2.csv')
df3=pd.read_csv('courses.csv')
nov=df2.copy()
dec=df2.copy()
```

```python
df1.head(1)
```

```python
df2.head(1)
```

```python
df1.merge(df2,how='inner',on='student_id')
```

```python
df1.merge(df2,how='left',on='student_id')
```

```python
df2.merge(df3,how='right',on='course_id')
```

```python
df2.merge(df3,how='outer',on='course_id').head(50) # it is like full outer join
```
